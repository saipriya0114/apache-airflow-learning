BashOperator with DAG


```python
from airflow import DAG
```

```python
from airflow.providers.standard.operators.bash import BashOperator
from datetime import datetime
```

```python
def extract():
  print("extracting")
def transform():
  print("transforming")
def load():
  print("loading")
```

```python
with DAG(
    dag_id='bash_ex',
    schedule='@hourly',
    start_date=datetime(2026,5,14),
    catchup=False
) as dag:
    task1 = BashOperator(
        task_id='extract',
        bash_command='echo "extracting"' # Changed to a string command for BashOperator
    )
    task2= BashOperator(
        task_id='transform',
        bash_command='echo "transforming"' # Changed to a string command for BashOperator
    )
    task3=BashOperator(
        task_id='load',
        bash_command='echo "loading"' # Changed to a string command for BashOperator
    )
```

```python
task1>>task2>>task3
```

```python
def fail_task():
  raise Exception("failed")
```

```python
with DAG(
    dag_id='bash_fail',
    schedule='@hourly',
    start_date=datetime(2026,5,14),
    catchup= False
) as dag:
    task1 = BashOperator(
        task_id='task_fail',
        bash_command=fail_task
    )
```

```python
with DAG(
    dag_id='bash_example',
    start_date=datetime(2025, 1, 1),
    schedule='@daily',
    catchup=False
) as dag:

    task1 = BashOperator(
        task_id='print_date',
        bash_command='date'
    )

    task2 = BashOperator(
        task_id='list_files',
        bash_command='ls'
    )

```
