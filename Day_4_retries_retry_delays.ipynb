```python
from airflow import DAG
from airflow.providers.standard.operators.python import PythonOperator
from datetime import datetime
from datetime import timedelta
```

```python
def my_date():
  print("Today's date")
```

```python
def extracting():
  print('extracting the file')
```

```python
def intentional_fail():
  print("intentionally failed")
```

```python
with DAG(
    dag_id='Day4_retries',
    schedule='@hourly',
    start_date=datetime(2026,5,15),
    catchup = False

    ) as dag:
        task1 = PythonOperator(
            task_id='date',
            python_callable=my_date
        )
        task2=PythonOperator(
            task_id='extracting',
            python_callable=extracting
        )
        task3=PythonOperator(
            task_id='fail',
            python_callable=intentional_fail,
            retries =3,
            retry_delay = timedelta(minutes=2)
        )
```

```python
task1>>task2>>task3
```

```python
from airflow import DAG
from airflow.providers.standard.operators.python import PythonOperator
from datetime import datetime
from datetime import timedelta

def print_date():
    print(f"current date and time: {datetime.now()}")

def extract():
    print("Extracting employee data...")



def transform():
    print("Transforming employee data...")

with DAG(
    dag_id='retries',
    start_date=datetime(2026, 6, 19),
    schedule='@hourly',
    catchup=False,
    ) as dag:

    print_date_task =PythonOperator(
        task_id='print_date',
        python_callable=print_date
    )

    extract_task = PythonOperator(
        task_id='extract',
        python_callable=extract
    )

    transform_task = PythonOperator(
        task_id = 'transform',
        python_callable =  transform,
        retries = 3,
        retry_delay =  timedelta(seconds=10)
    )
```

```python
print_date_task >> extract_task >> transform_task
```
